# Feature Selection
-Forward selection

```python
from statistics import linear_regression
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
import pandas as pd
df = pd.read_csv('Datasets/Cars93.csv')
df.isna().sum()
```

### Seperate X and y features

```python
X = df.drop(columns=['Id','Weight'])
Y = df[['Weight']]
X.head()
```

```python
Y.head()
```

```python
from PM6func import catconsep
cat, con = catconsep(X)
cat
```

```python
con
```

```python
X[con]
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OrdinalEncoder
```

```python
num_pipeline = Pipeline(steps=[('impute',SimpleImputer(strategy='mean')),
                               ('scaler',StandardScaler())])
cat_pipeline = Pipeline(steps=[('impute',SimpleImputer(strategy='most_frequent')),
                               ('Ordinal',OrdinalEncoder())])
pre = ColumnTransformer([('num',num_pipeline,con),
                         ('cat',cat_pipeline,cat)])
```

```python
X_pre = pre.fit_transform(X)
X_pre
```

```python
cols = pre.get_feature_names_out()
cols
```

```python
X_pre = pd.DataFrame(X_pre,columns=[cols])
X_pre.head()
```

### Feature Selection Forward with SequentialFeatureSelector

```python
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
sel = SequentialFeatureSelector(model, direction='forward')
sel_features = sel.fit_transform(X_pre, Y)
sel_cols = sel.get_feature_names_out(cols)
sel_cols
```

```python
sel_cols[0]
```

```python
sel_cols[0].split('__')
```

```python
imp_cols = []
for col in sel_cols:
    imp_cols.append(col.split('__')[1])
imp_cols
```

# Feature Selection in X

```python
X_sel = X[imp_cols]
X_sel.head()
```

```python
xselcat, xselcon = catconsep(X_sel)
xselcat
```

```python
xselcon
```

```python
X_sel[xselcon]
```

## Create Pipeline for X selected Feature

```python
num_pipeline2 = Pipeline(steps=[('impute',SimpleImputer(strategy='mean')),
                                ('scaler',StandardScaler())])
cat_pipeline2 = Pipeline(steps=[('impute',SimpleImputer(strategy='most_frequent')),
                                ('ordinal',OrdinalEncoder())])
X_pre2 = ColumnTransformer([('num',num_pipeline2,xselcon),
                            ('cat',cat_pipeline2,xselcat)])
X_pre2
```

```python
x_sel_pre = X_pre2.fit_transform(X_sel)
x_sel_pre
```

```python
final_cols = X_pre2.get_feature_names_out()
final_cols
```

```python
X_final = pd.DataFrame(x_sel_pre,columns=[final_cols])
X_final.head()
```

### Train Test Split
1. 80% train data
2. 20% test data

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X_final, Y, test_size=0.2, random_state=21)
```
