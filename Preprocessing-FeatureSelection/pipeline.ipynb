# Pipeline
### Automate the preprocessing 

```python
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
```

```python
import pandas as pd
df = pd.read_csv("Datasets\Cars93.csv")
df.head()
```

```python
X =df.drop(columns=['Id','Weight'])
Y = df[['Weight']]
```

```python
from PM6func import catconsep
X_cat,X_con = catconsep(X)
```

```python
X_con
```

```python
num_pipeline = Pipeline(steps=[('impute',SimpleImputer(strategy='mean')),
                              ('scaler',StandardScaler())])
cat_pipeline = Pipeline(steps=[('impute',SimpleImputer(strategy=('most_frequent'))),
                               ('OHE',OneHotEncoder(handle_unknown='ignore'))])
```

```python
preprocessor = ColumnTransformer([('num',num_pipeline,X_con),
                                  ('cat',cat_pipeline,X_cat)])
```

```python
X_pre = preprocessor.fit_transform(X).toarray()
X_pre
```

```python
cols = preprocessor.get_feature_names_out()
cols
```

```python
X_pre = pd.DataFrame(X_pre, columns=[cols])
X_pre.head()
```

### Train test split
1. 80% data train to model
2. 20% data not pass to the model

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X_pre,Y, test_size=0.2, random_state=21)
```
