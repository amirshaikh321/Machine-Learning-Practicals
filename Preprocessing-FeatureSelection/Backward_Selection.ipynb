# Feature Selection
### Backward Selection

```python
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
import pandas as pd
df = pd.read_csv('Datasets/Cars93.csv')
df.head()
```

### Seperate x and y columns

```python
X = df.drop(columns=['Id', 'Weight'])
Y = df[['Weight']]
```

```python
X.head()
```

```python
Y.head()
```

Seperate cat and con columns

```python
from PM6func import catconsep
cat, con = catconsep(X)
```

Create Pipeline For preprocessing data

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OrdinalEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
num_pipeline1 = Pipeline(steps=[('impute', SimpleImputer(strategy='mean')),
                                ('scaler',StandardScaler())])
cat_pipeline1 = Pipeline(steps=[('impuute',SimpleImputer(strategy='most_frequent')),
                                ('ordinal',OrdinalEncoder())])
pre= ColumnTransformer([('num',num_pipeline1,con),
                       ('cat',cat_pipeline1,cat)])
```

```python
pre
```

```python
X_pre = pre.fit_transform(X)
X_pre
```

```python
cols = pre.get_feature_names_out()
cols
```

```python
X_pre = pd.DataFrame(X_pre, columns=[cols])
X_pre.head()
```

### Feature selection
-Backward Selection

```python
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import SequentialFeatureSelector

model = LinearRegression()
sel = SequentialFeatureSelector(model, direction='backward')
sel_feature = sel.fit_transform(X_pre,Y)
sel_cols = sel.get_feature_names_out(cols)
sel_cols
```

get original columnn names

```python
imp_cols=[]
for i in sel_cols:
    s=i.split('__')[1]
    imp_cols.append(s)
```

```python
imp_cols
```

```python
X_pre_sel = X[imp_cols]
```

```python
X_pre_sel.head()
```

```python
xselcat, xselcon = catconsep(X_pre_sel)
```

### Now again Preprocess This new dataframe data

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
num_pipeline2 = Pipeline(steps=[('impute',SimpleImputer(strategy='mean')),
                                ('scaler',StandardScaler())])
cat_pipeline2 = Pipeline(steps=[('impute',SimpleImputer(strategy='most_frequent')),
                                ('ohe',OneHotEncoder())])
pre2 = ColumnTransformer([('num',num_pipeline2,xselcon),
                          ('cat',cat_pipeline2,xselcat)])
```

```python
X_final = pre2.fit_transform(X_pre_sel).toarray()
X_final
```

```python
final_cols = pre2.get_feature_names_out()
final_cols
```

```python
X_final = pd.DataFrame(X_final, columns=[final_cols])
X_final.head()
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X_final,Y, test_size=0.2, random_state=21)
xtrain.head()
```

```python
ytrain.head()
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(xtrain,ytrain)
```

```python
ypred_tr = model.predict(xtrain)
ypred_tr[0:5]
```

### r2score of train data
