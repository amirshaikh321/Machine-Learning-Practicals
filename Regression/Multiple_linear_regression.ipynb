# Multiple Linear Regression

Income ~ Age, Experience

```python
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
df = pd.read_csv('Datasets\multiple_linear_regression_dataset.csv')
df.head()
```

```python
X = df.drop(columns=['income'])
X.head()
```

```python
Y = df[['income']]
Y.head()
```

here income ~ experience

```python
plt.scatter(df['experience'],df['income'])
plt.show()
```

### Train Model

```python
model = LinearRegression()
model.fit(X, Y)
```

### Coeficient and Intercept

```python
model.coef_
```

```python
model.intercept_
```

### Income ~ Age, Experience
income = B0 + B1*age +B2*experience
1. B0 = 31261.6898
2. B1 = -99.1953
3. B2 = 2162.4041

```python
y_pred = model.predict(X)
y_pred
```

```python
Y.head()
```

### Evaluate based on different metrics
1. Mean Squared Error
2. Mean Absolute Error
3. Root  Mean Squared Error
4. R2Score

```python
mse = mean_squared_error(Y,y_pred)
mse
```

```python
mae = mean_squared_error(Y,y_pred)
mae
```

```python
r2 = r2_score(Y,y_pred)
r2
```
